```python
table2=pd.read_csv("report_07-27.csv")
```

```python
df = pd.concat([table, table1,table2], ignore_index=True)

# Upewnij się, że time_stamp to datetime
df["time_stamp"] = pd.to_datetime(df["time_stamp"])

# Wyodrębnij datę + godzinę (bez minut i sekund)
df["datetime_hour"] = df["time_stamp"].dt.floor("H")

# Wybierz potrzebne kolumny
df_filtered = df[["number", "direction", "stopName", "plannedTime", "delay", "datetime_hour"]]

# Grupowanie i liczenie średniej wartości opóźnienia
result = (
    df_filtered
    .groupby(["number", "direction", "stopName", "datetime_hour", "plannedTime"], as_index=False)
    .agg({"delay": "mean"})
)

# Zmień nazwy kolumn na żądane
result.rename(columns={
    "number": "busLineNumber",
    "stopName": "bus stop",
    "delay": "avg_delay"
}, inplace=True)

print(result.head())
```
